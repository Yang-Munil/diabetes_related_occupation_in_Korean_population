### 데이터 불러오기

```python
df_01 = pd.read_csv("./merge_20201207_target_12.csv")
df_02 = pd.read_csv("./merge_20210113.csv")
```

### 타겟값 만들기 (12월)

```python
### 변수 추가

df_01['A02_CNP_AMT_BUG'] = df_01['A02_CNP_AMT']/df_01['A01_BUG_AMT_3Y']
df_01['A02_CNP_AMT_BUG'] = df_01['A02_CNP_AMT_BUG'].fillna(0)

df_02['A02_CNP_AMT_BUG'] = df_02['A02_CNP_AMT']/df_02['A01_BUG_AMT_3Y']
df_02['A02_CNP_AMT_BUG'] = df_02['A02_CNP_AMT_BUG'].fillna(0)


print('dataset shape:', df_01.shape)
print('dataset shape:', df_02.shape)

### 다음 달 데이터에 TARGET 이 0인 컬럼 추가후 비식별코드/타겟 컬럼만 뽑아내기

df_02['TARGET_01'] = 0

display(df_02)

df_02 = df_02[['TPR_NO_ENC' ,'TARGET_01']]

df_02.head()

# 전월 데이터에 반영

left_outer_join_result = df_01.merge(df_02, on='TPR_NO_ENC', how='left')
left_outer_join_result

left_outer_join_result['TARGET_01'] = left_outer_join_result['TARGET_01'].fillna(1)

left_outer_join_result['TARGET_01'] = left_outer_join_result['TARGET_01'].astype(int)

display(left_outer_join_result)
```

```python
# 체납자, 수납자 구분을 위한 새로운 컬럼 생성.
# 0 = 체납자 / 1,2 = 수납자

left_outer_join_result['TARGET'] = left_outer_join_result['TARGET_12'] + left_outer_join_result['TARGET_01']
```

```python
# 체납자

nopay = left_outer_join_result.loc[left_outer_join_result['TARGET']==0]
nopay.shape
```

```python
# 수납자

yespay = left_outer_join_result.loc[(left_outer_join_result['TARGET']==1) | (left_outer_join_result['TARGET']==2)]
yespay.shape
```

```python
nopay.to_csv('체납자_11.csv', index=False)
```

```python
yespay.to_csv('수납자_11.csv', index=False)
```

```python
import gc

del left_outer_join_result
del df_01
del df_02
del nopay
del yespay

gc.collect()
```
